```python
users=[]
```

```python
def getAuxStr(arr):
    aux=""
    for i in arr:
        aux=aux+" "+str(i)
    return aux.strip()
```

```python
getAuxStr(['Uddeshya','Singh','Rocks'])
```

```python
def findUser(arr):
    for i in range(2,len(arr)):
        if arr[i][-1]==":":
            aux_string=getAuxStr(arr[3:i])
            users.append((aux_string+" "+arr[i].split(':')[0]).strip())
```

```python
for i in aux:
    findUser(i.split())
```

```python
print users
```

```python
for i in set(users):
    print i
```

```python
threshold=10
verified=[]
for i in set(users):
    count=0
    for j in users:
        if i==j:
            count+=1
    if count>threshold:
        verified.append(i)
    
```
